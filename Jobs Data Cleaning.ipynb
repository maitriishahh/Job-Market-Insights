```python
import ast

# Safely extract sector/industry
def extract_value(x, key):
    try:
        if isinstance(x, str):   # only parse if it's a string
            return ast.literal_eval(x).get(key)
    except:
        return None
    return None

df["Sector"] = df["Company_Profile"].apply(lambda x: extract_value(x, "Sector"))
df["Industry"] = df["Company_Profile"].apply(lambda x: extract_value(x, "Industry"))
```

```python
df['Experience'] = df['Experience'].str.replace("to","-",regex=False)
```

```python
import re

# Assume your DataFrame is loaded and stored in a variable named 'df'.
# For demonstration purposes, a sample DataFrame is created below.
# You will replace this part with your actual code to load your data (e.g., from a CSV file).

# Create an empty list to store the cleaned skills.
cleaned_skills_list = []

# Iterate over each skill string in the original 'Skills' column.
for skill_string in df1['Skills']:
    # Step 1: Remove parenthetical notes.
    cleaned_string = re.sub(r'\s*\(.*?\)', '', skill_string)

    # Step 2: Normalize multiple spaces to a single space.
    cleaned_string = re.sub(r'\s+', ' ', cleaned_string)

    # Step 3: Insert a comma before each capitalized word to create a consistent delimiter.
    # The regex `\s([A-Z])` finds a space followed by an uppercase letter.
    cleaned_string = re.sub(r'\s([A-Z])', r', \1', cleaned_string)

    # Step 4: Split the string by the comma delimiter and clean up each skill.
    individual_skills = [skill.strip() for skill in cleaned_string.split(',')]

    # Step 5: Filter out any empty strings that may have resulted from the split.
    final_skills = [skill for skill in individual_skills if skill]

    # Add the list of cleaned skills to our list.
    cleaned_skills_list.append(final_skills)

# Add the new list of cleaned skills to the DataFrame as a new column.
df1['Cleaned_Skills'] = cleaned_skills_list

# Display the DataFrame with both original and cleaned skills columns.
print(df1[['Skills', 'Cleaned_Skills']])
```
